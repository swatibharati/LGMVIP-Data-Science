```python
import cv2
import io
from PIL import Image
```

```python
#Read Image
img=cv2.imread("dog.jpeg")
img=cv2.resize(img,(500,650))
display(Image.fromarray(img))
```

```python
#CONVERSION OF ORIGINAL IMAGE TO GRAY_SCALE
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_img))
```

```python
#CONVERSION OF GRAY_SCALE TO INVERTED_IMAGE
invert_img=cv2.bitwise_not(gray_img)
display(Image.fromarray(invert_img))
```

```python
invert_img=cv2.bitwise_not(gray_img)
display(Image.fromarray(invert_img))
```

```python
#Blured Image
blur=cv2.GaussianBlur(invert_img,(5,5),0,0)
display(Image.fromarray(blur))
```

```python
#PENCIL_SKETCH
invert_blur=255-blur
pencil_sketch=cv2.divide(gray_img,invert_blur,scale=250.0)
display(Image.fromarray(pencil_sketch))
```

```python
display(Image.fromarray(img))
display(Image.fromarray(pencil_sketch))
```
